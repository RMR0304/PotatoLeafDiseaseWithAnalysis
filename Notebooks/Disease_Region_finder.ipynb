```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import cv2
from glob import glob
import random
import warnings
warnings.filterwarnings('ignore')

try:
    import segmentation_models_pytorch as smp
except ImportError:
    %pip install segmentation_models_pytorch
    import segmentation_models_pytorch as smp

try:
    import albumentations as A
    from albumentations.pytorch import ToTensorV2
except ImportError:
    %pip install albumentations
    import albumentations as A
    from albumentations.pytorch import ToTensorV2

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as T

from sklearn.model_selection import train_test_split
from torch.optim.lr_scheduler import ReduceLROnPlateau, CosineAnnealingLR
import torch.nn.functional as F

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name()}")
    print(f"Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")

from google.colab import drive
drive.mount('/content/drive')

DATASET_PATH = '/content/drive/MyDrive/aug_data'  

class DataExplorer:
    def __init__(self, dataset_path):
        self.images_path = os.path.join(dataset_path, 'images')
        self.masks_path = os.path.join(dataset_path, 'masks')
        self.image_paths = sorted(glob(os.path.join(self.images_path, '*.jpg')) +
                                 glob(os.path.join(self.images_path, '*.png')))
        self.mask_paths = sorted(glob(os.path.join(self.masks_path, '*.png')) +
                                glob(os.path.join(self.masks_path, '*.jpg')))

    def explore(self):
        print(f"Total images: {len(self.image_paths)}")
        print(f"Total masks: {len(self.mask_paths)}")

        assert len(self.image_paths) == len(self.mask_paths), "Image and mask counts don't match!"

        fig, axes = plt.subplots(2, 5, figsize=(15, 6))
        for i in range(5):
            idx = random.randint(0, len(self.image_paths)-1)

            img = cv2.imread(self.image_paths[idx])
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            mask = cv2.imread(self.mask_paths[idx], cv2.IMREAD_GRAYSCALE)

            axes[0, i].imshow(img)
            axes[0, i].set_title(f'Image {idx}')
            axes[0, i].axis('off')

            axes[1, i].imshow(mask, cmap='gray')
            axes[1, i].set_title(f'Mask {idx}')
            axes[1, i].axis('off')

        plt.tight_layout()
        plt.show()

        disease_areas = []
        for mask_path in self.mask_paths[:100]:
            mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
            disease_ratio = (mask > 0).sum() / mask.size
            disease_areas.append(disease_ratio)

        plt.figure(figsize=(10, 5))
        plt.hist(disease_areas, bins=50, alpha=0.7, color='blue')
        plt.xlabel('Disease Area Ratio')
        plt.ylabel('Frequency')
        plt.title('Distribution of Disease Area in Images')
        plt.show()

        return self.image_paths, self.mask_paths

class LeafDiseaseDataset(Dataset):
    def __init__(self, image_paths, mask_paths, transform=None, augment=False):
        self.image_paths = image_paths
        self.mask_paths = mask_paths
        self.transform = transform
        self.augment = augment

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image = cv2.imread(self.image_paths[idx])
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        mask = cv2.imread(self.mask_paths[idx], cv2.IMREAD_GRAYSCALE)

        mask = (mask > 0).astype(np.float32)

        if self.transform:
            augmented = self.transform(image=image, mask=mask)
            image = augmented['image']
            mask = augmented['mask']

        if len(mask.shape) == 2:
            mask = np.expand_dims(mask, axis=0)

        return torch.FloatTensor(image), torch.FloatTensor(mask)

def get_transforms(img_size=(256, 256)):
    train_transform = A.Compose([
        A.Resize(img_size[0], img_size[1]),
        A.RandomRotate90(p=0.5),
        A.HorizontalFlip(p=0.5),
        A.ShiftScaleRotate(shift_limit=0.0625, scale_limit=0.1, rotate_limit=15, p=0.5),
        A.OneOf([
            A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=1),
            A.HueSaturationValue(hue_shift_limit=20, sat_shift_limit=30, val_shift_limit=20, p=1),
        ], p=0.7),
        A.OneOf([
            A.GaussNoise(var_limit=(10.0, 50.0), p=1),
            A.GaussianBlur(blur_limit=(3, 5), p=1),
            A.CLAHE(clip_limit=2.0, tile_grid_size=(8, 8), p=1),
        ], p=0.5),
        A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
        ToTensorV2(),
    ])

    val_transform = A.Compose([
        A.Resize(img_size[0], img_size[1]),
        A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
        ToTensorV2(),
    ])

    return train_transform, val_transform

class LeafDiseaseSegmentationModel:
    def __init__(self, encoder_name='efficientnet-b3', encoder_weights='imagenet',
                 in_channels=3, classes=1, device='cuda'):
        self.device = device
        self.model = smp.Unet(
            encoder_name=encoder_name,
            encoder_weights=encoder_weights,
            in_channels=in_channels,
            classes=classes,
            activation=None, 
        ).to(device)

        self.bce_loss = nn.BCEWithLogitsLoss()
        self.dice_loss = smp.losses.DiceLoss(mode='binary')

    def forward(self, x):
        return self.model(x)

    def compute_loss(self, outputs, targets):
        bce = self.bce_loss(outputs, targets)
        dice = self.dice_loss(outputs, targets)
        return bce + dice

    def calculate_metrics(self, outputs, targets, threshold=0.5):
        probs = torch.sigmoid(outputs)
        preds = (probs > threshold).float()

        tp = (preds * targets).sum()
        fp = (preds * (1 - targets)).sum()
        fn = ((1 - preds) * targets).sum()
        tn = ((1 - preds) * (1 - targets)).sum()

        iou = tp / (tp + fp + fn + 1e-7)

        dice = 2 * tp / (2 * tp + fp + fn + 1e-7)

        precision = tp / (tp + fp + 1e-7)
        recall = tp / (tp + fn + 1e-7)

        return {
            'iou': iou.item(),
            'dice': dice.item(),
            'precision': precision.item(),
            'recall': recall.item()
        }

class Trainer:
    def __init__(self, model, train_loader, val_loader, optimizer, scheduler, device):
        self.model = model
        self.train_loader = train_loader
        self.val_loader = val_loader
        self.optimizer = optimizer
        self.scheduler = scheduler
        self.device = device

        self.train_losses = []
        self.val_losses = []
        self.val_metrics = []
        self.best_val_loss = float('inf')

    def train_epoch(self):
        self.model.model.train() # Corrected: Call train() on the underlying PyTorch model
        total_loss = 0

        for batch_idx, (images, masks) in enumerate(tqdm(self.train_loader, desc="Training")):
            images = images.to(self.device)
            masks = masks.to(self.device)

            self.optimizer.zero_grad()
            outputs = self.model.forward(images) # Corrected: Call forward() on the LeafDiseaseSegmentationModel instance
            loss = self.model.compute_loss(outputs, masks)
            loss.backward()
            self.optimizer.step()

            total_loss += loss.item()

        avg_loss = total_loss / len(self.train_loader)
        self.train_losses.append(avg_loss)
        return avg_loss

    def validate(self):
        self.model.model.eval() # Corrected: Call eval() on the underlying PyTorch model
        total_loss = 0
        all_metrics = []

        with torch.no_grad():
            for images, masks in tqdm(self.val_loader, desc="Validating"):
                images = images.to(self.device)
                masks = masks.to(self.device)

                outputs = self.model.forward(images) # Corrected: Call forward() on the LeafDiseaseSegmentationModel instance
                loss = self.model.compute_loss(outputs, masks)
                total_loss += loss.item()

                metrics = self.model.calculate_metrics(outputs, masks)
                all_metrics.append(metrics)

        avg_loss = total_loss / len(self.val_loader)
        self.val_losses.append(avg_loss)

        # Average metrics
        avg_metrics = {key: np.mean([m[key] for m in all_metrics]) for key in all_metrics[0].keys()}
        self.val_metrics.append(avg_metrics)

        return avg_loss, avg_metrics

    def train(self, epochs):
        for epoch in range(epochs):
            print(f"\nEpoch {epoch+1}/{epochs}")
            print("-" * 50)

            # Train
            train_loss = self.train_epoch()
            print(f"Train Loss: {train_loss:.4f}")

            # Validate
            val_loss, val_metrics = self.validate()
            print(f"Val Loss: {val_loss:.4f}")
            print(f"Val Metrics - IoU: {val_metrics['iou']:.4f}, Dice: {val_metrics['dice']:.4f}, "
                  f"Precision: {val_metrics['precision']:.4f}, Recall: {val_metrics['recall']:.4f}")

            # Learning rate scheduling
            if self.scheduler:
                if isinstance(self.scheduler, ReduceLROnPlateau):
                    self.scheduler.step(val_loss)
                else:
                    self.scheduler.step()

            # Save best model
            if val_loss < self.best_val_loss:
                self.best_val_loss = val_loss
                torch.save(self.model.model.state_dict(), 'best_model.pth')
                print(f"✓ Model saved! (Val Loss: {val_loss:.4f})")

        return self.train_losses, self.val_losses, self.val_metrics

    def plot_training_history(self):
        fig, axes = plt.subplots(1, 3, figsize=(15, 4))

        # Loss plot
        axes[0].plot(self.train_losses, label='Train Loss')
        axes[0].plot(self.val_losses, label='Val Loss')
        axes[0].set_xlabel('Epoch')
        axes[0].set_ylabel('Loss')
        axes[0].set_title('Training and Validation Loss')
        axes[0].legend()
        axes[0].grid(True)

        # Metrics plot
        epochs = range(1, len(self.val_metrics) + 1)
        iou = [m['iou'] for m in self.val_metrics]
        dice = [m['dice'] for m in self.val_metrics]

        axes[1].plot(epochs, iou, label='IoU')
        axes[1].plot(epochs, dice, label='Dice')
        axes[1].set_xlabel('Epoch')
        axes[1].set_ylabel('Score')
        axes[1].set_title('Validation Metrics')
        axes[1].legend()
        axes[1].grid(True)

        # Precision-Recall plot
        precision = [m['precision'] for m in self.val_metrics]
        recall = [m['recall'] for m in self.val_metrics]

        axes[2].plot(epochs, precision, label='Precision')
        axes[2].plot(epochs, recall, label='Recall')
        axes[2].set_xlabel('Epoch')
        axes[2].set_ylabel('Score')
        axes[2].set_title('Precision and Recall')
        axes[2].legend()
        axes[2].grid(True)

        plt.tight_layout()
        plt.show()

# ============================================
# SECTION 8: PREDICTION AND VISUALIZATION
# ============================================

class Predictor:
    def __init__(self, model_path, model_class, device='cuda'):
        self.device = device
        self.model = model_class.model
        self.model.load_state_dict(torch.load(model_path))
        self.model.to(device)
        self.model.eval()

    def predict_single(self, image, threshold=0.5):
        """Predict mask for a single image"""
        # Preprocess
        transform = A.Compose([
            A.Resize(256, 256),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2(),
        ])

        augmented = transform(image=image)
        input_tensor = augmented['image'].unsqueeze(0).to(self.device)

        # Predict
        with torch.no_grad():
            output = self.model(input_tensor)
            prob = torch.sigmoid(output)
            pred = (prob > threshold).float()

        # Resize back to original size
        pred = pred.squeeze().cpu().numpy()
        pred = cv2.resize(pred, (image.shape[1], image.shape[0]))

        return pred

    def visualize_predictions(self, image_paths, mask_paths, num_samples=5):
        """Visualize predictions alongside ground truth"""
        fig, axes = plt.subplots(num_samples, 3, figsize=(12, 4*num_samples))

        for i in range(num_samples):
            idx = random.randint(0, len(image_paths)-1)

            # Load image and ground truth
            image = cv2.imread(image_paths[idx])
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            gt_mask = cv2.imread(mask_paths[idx], cv2.IMREAD_GRAYSCALE)
            gt_mask = (gt_mask > 0).astype(np.uint8)

            # Predict
            pred_mask = self.predict_single(image)

            # Overlay visualization
            overlay = image.copy()
            overlay[pred_mask > 0] = [255, 0, 0]  # Red for predicted disease

            # Plot
            axes[i, 0].imshow(image)
            axes[i, 0].set_title('Original Image')
            axes[i, 0].axis('off')

            axes[i, 1].imshow(gt_mask, cmap='gray')
            axes[i, 1].set_title('Ground Truth')
            axes[i, 1].axis('off')

            axes[i, 2].imshow(overlay)
            axes[i, 2].set_title('Prediction (Red)')
            axes[i, 2].axis('off')

        plt.tight_layout()
        plt.show()

# ============================================
# SECTION 9: MAIN EXECUTION PIPELINE
# ============================================

def main():
    # Set parameters
    IMG_SIZE = (256, 256)
    BATCH_SIZE = 16
    EPOCHS = 20
    LEARNING_RATE = 1e-3
    VAL_SPLIT = 0.2
    RANDOM_SEED = 42

    # Set random seeds for reproducibility
    torch.manual_seed(RANDOM_SEED)
    np.random.seed(RANDOM_SEED)
    random.seed(RANDOM_SEED)

    # Step 1: Explore data
    print("Step 1: Exploring dataset...")
    explorer = DataExplorer(DATASET_PATH)
    image_paths, mask_paths = explorer.explore()

    # Step 2: Split data
    print("\nStep 2: Splitting dataset...")
    train_images, val_images, train_masks, val_masks = train_test_split(
        image_paths, mask_paths, test_size=VAL_SPLIT, random_state=RANDOM_SEED
    )
    print(f"Training samples: {len(train_images)}")
    print(f"Validation samples: {len(val_images)}")

    # Step 3: Create data loaders
    print("\nStep 3: Creating data loaders...")
    train_transform, val_transform = get_transforms(IMG_SIZE)

    train_dataset = LeafDiseaseDataset(train_images, train_masks, transform=train_transform)
    val_dataset = LeafDiseaseDataset(val_images, val_masks, transform=val_transform)

    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)

    # Step 4: Initialize model
    print("\nStep 4: Initializing model...")
    model = LeafDiseaseSegmentationModel(
        encoder_name='efficientnet-b3',
        encoder_weights='imagenet',
        in_channels=3,
        classes=1,
        device=device
    )

    # Step 5: Setup optimizer and scheduler
    optimizer = optim.AdamW(model.model.parameters(), lr=LEARNING_RATE, weight_decay=1e-4)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5)

    # Step 6: Train model
    print("\nStep 5: Starting training...")
    trainer = Trainer(model, train_loader, val_loader, optimizer, scheduler, device)
    train_losses, val_losses, val_metrics = trainer.train(EPOCHS)

    # Step 7: Plot training history
    print("\nStep 6: Plotting training history...")
    trainer.plot_training_history()

    # Step 8: Evaluate on validation set
    print("\nStep 7: Final evaluation...")
    final_val_loss, final_metrics = trainer.validate()
    print(f"Final Validation Results:")
    print(f"Loss: {final_val_loss:.4f}")
    print(f"IoU: {final_metrics['iou']:.4f}")
    print(f"Dice: {final_metrics['dice']:.4f}")
    print(f"Precision: {final_metrics['precision']:.4f}")
    print(f"Recall: {final_metrics['recall']:.4f}")

    # Step 9: Visualize predictions
    print("\nStep 8: Visualizing predictions...")
    predictor = Predictor('best_model.pth', model, device)
    predictor.visualize_predictions(val_images, val_masks, num_samples=5)

    print("\n✓ Training completed successfully!")

# ============================================
# SECTION 10: RUN THE PIPELINE
# ============================================

if __name__ == "__main__":
    main()
```

```python
import cv2
import numpy as np
import albumentations as A
from albumentations.pytorch import ToTensorV2

# Transform (same as validation)
transform = A.Compose([
    A.Resize(256, 256),
    A.Normalize(mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225]),
    ToTensorV2(),
])

def predict_image(image_path, threshold=0.5):
    # Read image
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    orig = image.copy()

    # Transform
    augmented = transform(image=image)
    input_tensor = augmented['image'].unsqueeze(0).to(device)

    # Predict
    with torch.no_grad():
        output = model(input_tensor)
        prob = torch.sigmoid(output)
        mask = (prob > threshold).float()

    # Convert mask
    mask = mask.squeeze().cpu().numpy()
    mask = cv2.resize(mask, (orig.shape[1], orig.shape[0]))

    return orig, mask
```
